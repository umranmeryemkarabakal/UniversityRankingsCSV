```python
import pandas as pd
import numpy as np
import matplotlib as plt
from collections import Counter
```

```python
dataFrame = pd.read_csv('University_Rankings_20220204.csv')
```

```python
dataFrame.head(3)
```

```python
dataFrame.rename(columns={'world_rank': 'dünya_sıralaması',
'university_name': 'üniversite_adı',
'country': 'ülke',
'teaching': 'öğretim',
'international': 'uluslararası',
'research': 'araştırma',
'citations': 'atıflar',
'income': 'gelir',
'total_score': 'toplam_puan',
'num_students': 'öğrenci_sayısı',
'student_staff_ratio': 'öğrenci_personel_oranı',
'international_students': 'uluslararası_öğrenciler',
'female_male_ratio': 'kadın_erkek_oranı',
'year': 'yıl'
},inplace=True)
```

```python
dataFrame.isnull().sum()
```

```python
dataFrame["öğrenci_sayısı"] = dataFrame["öğrenci_sayısı"].str.replace(',', '', regex=False).str.strip().astype('Int64')
```

```python
dataFrame = dataFrame[~pd.isnull(dataFrame["uluslararası_öğrenciler"])]
```

```python
dataFrame["uluslararası_öğrenciler"] = dataFrame["uluslararası_öğrenciler"].str.replace('%', '', regex=False).str.strip().astype(float) / 100
```

```python
dataFrame.info()
```

```python
dataFrame.isnull().sum()
```

```python
dataFrame["kadın_erkek_oranı"] = dataFrame["kadın_erkek_oranı"].str.replace('-', ':', regex=False).str.strip()
dataFrame["kadın_erkek_oranı"] = dataFrame["kadın_erkek_oranı"].str.replace(' ', '', regex=False).str.strip()
```

```python
dataFrame["kadın_erkek_oranı"].unique()
```

```python
dataFrame['kadın_erkek_oranı'] = dataFrame['kadın_erkek_oranı'].replace(':', pd.NA)
```

```python
dataFrame["kadın_erkek_oranı"]=dataFrame["kadın_erkek_oranı"].fillna("50:50")
```

```python
dataFrame[['kadın_oranı', 'erkek_oranı']] = dataFrame['kadın_erkek_oranı'].str.strip().str.split(':', expand=True).astype(float) /100
```

```python
dataFrame.drop(["kadın_erkek_oranı"],axis=1,inplace=True)
```

```python
dataFrame.info()
```

```python
(dataFrame["toplam_puan"]=="-").sum()
```

```python
headList = ["toplam_puan","gelir", "uluslararası"]
for i in headList:
    dataFrame[i] = dataFrame[i].replace('-', pd.NA)
    dataFrame[i].dropna().astype(float).describe()
    dataFrame[i] = dataFrame[i].fillna(str(dataFrame[i].dropna().astype(float).mean()))
    dataFrame[i] = dataFrame[i].astype(float)
```

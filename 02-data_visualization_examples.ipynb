```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataFrame = pd.read_csv("university_rankings_temizlenmis.csv")
```

```python
dataFrame["ülke"].unique()
```

```python
kategorikDataFrame = dataFrame[dataFrame['ülke'].isin(dataFrame['ülke'].value_counts().head(3).index)]
kategorikDataFrame.head(2)
```

```python
kategorikDataFrame["ülke"].value_counts()
```

```python
#scater plot

sns.set_style("darkgrid")
sns.scatterplot(data=dataFrame,x="araştırma",y="öğretim")
plt.title("araştırma puanı - öğretim puanı ilişkisi")
plt.show()
```

```python
#kategorik scater plot

sns.set_style("darkgrid")
sns.scatterplot(data= kategorikDataFrame ,x="araştırma",y="öğretim",hue = "ülke",palette='Set1')
plt.title("araştırma puanı - öğretim puanı ilişkisi")
plt.show()
```

```python
#doğrusal regresyon

sns.regplot(x="araştırma",y="öğretim",data=dataFrame,marker='*',scatter_kws={'alpha':0.2})
plt.title("araştırma öğretim puanı dağılımı")
plt.show()
```

```python
#çizgi grafiği

sns.lineplot(x="araştırma",y="öğretim",data=dataFrame)
plt.title("araştırma puanı ve öğretim puanı")
plt.show()
```

```python
#kategorik çizgi grafiği

sns.lineplot(x="araştırma",y="öğretim",data=kategorikDataFrame,hue="ülke")
plt.title("araştırma puanı ve öğretim puanı")
plt.show()
```

```python
dataFrame.columns
```

```python
#histogram

sns.displot(dataFrame,x='öğrenci_sayısı')
plt.ylabel("frekans")
plt.title("öğrenci sayısı histogramı")
plt.show()
```

```python
#kategorik histogram

sns.displot(kategorikDataFrame ,x="öğrenci_sayısı",hue="ülke")
plt.ylabel("frekans")
plt.title("ülkeye göre ayrıştırılmış öğrenci sayısı histogramı")
plt.show()
```

```python
sns.displot(kategorikDataFrame,x="öğrenci_sayısı",col="ülke",multiple = 'dodge')
plt.show()
```

```python
sns.displot(kategorikDataFrame,x="araştırma",y="öğretim",hue = "ülke",kind="kde")
plt.xlabel("araştırma")
plt.ylabel("öğretim")
plt.title("ülkeye göre ayrıştırılmış araştırma öğretim puanları histogramı")
plt.show()
```

```python
sns.kdeplot(kategorikDataFrame,x="araştırma",y="öğretim",hue = "ülke",fill = True)
plt.xlabel("araştırma")
plt.ylabel("öğretim")
plt.title("ülkeye göre ayrıştırılmış araştırma öğretim puanları histogramı")
plt.show()
```

```python
#çubuk grafiği

sns.barplot(x = "ülke",y= "öğretim",data= kategorikDataFrame)
plt.title("ülkelere göre öğretim puanı grafikleri")
plt.show()
```

```python
#kategorik çubuk grafiği

sns.catplot(x="yıl",y="atıflar",col="ülke",data=kategorikDataFrame,kind="bar",height=4,aspect=0.7,palette='Set1')
plt.show()
```

```python
#box plot 

sns.boxplot(x="ülke",y="uluslararası",data=kategorikDataFrame)
plt.show()
```

```python
sns.catplot(x="yıl",y="uluslararası",col="ülke",data=kategorikDataFrame,kind="box",height=4,aspect=0.7,palette="rocket")
plt.show()
```

```python
kisitliDataFrame = dataFrame[dataFrame['ülke'].isin(dataFrame['ülke'].value_counts().head(35).index)]
```

```python
#sayma grafiği

sns.countplot(x="ülke",data=kisitliDataFrame)
plt.xticks(rotation=90)
plt.show()
```

```python
#ortaklık grafiği

sns.jointplot(data=kategorikDataFrame,x="öğretim",y="atıflar",hue="ülke",kind="kde")
plt.show()
```

```python
g= sns.JointGrid(data=dataFrame,x="öğretim",y="atıflar")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
plt.show()
```

```python
dataFrame_gecici = kategorikDataFrame.loc[:,["öğretim","uluslararası","araştırma","atıflar","gelir"]]
```

```python
#çift/eş grafiği

sns.pairplot(dataFrame_gecici)
plt.show()
```

```python
g= sns.PairGrid(dataFrame_gecici)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot,fill=True)
g.map_diag(sns.histplot,kde=True)
plt.show()
```
